```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from matplotlib.table import Table
import seaborn as sns
from matplotlib.patches import RegularPolygon

def create_table_with_heatmap_arrows(data, title=None, cmap='RdYlGn',
                                     arrow_column=None, figsize=(10, 6),
                                     font_size=12, cell_height=0.05):
    """
    Create a table with heatmap background colors and direction arrows.

    Parameters:
    -----------
    data : pandas DataFrame
        The data to visualize in the table.
    title : str, optional
        Title of the table.
    cmap : str or matplotlib colormap, optional
        Colormap to use for the heatmap.
    arrow_column : str or list, optional
        Column(s) to add direction arrows based on values.
    figsize : tuple, optional
        Figure size (width, height).
    font_size : int, optional
        Font size for the table text.
    cell_height : float, optional
        Height of each cell in the table.

    Returns:
    --------
    fig, ax : matplotlib figure and axis objects
    """
    # Create figure and axis
    fig, ax = plt.subplots(figsize=figsize)
    ax.axis('off')

    # If arrow_column is a string, convert to list
    if isinstance(arrow_column, str):
        arrow_column = [arrow_column]
    elif arrow_column is None:
        arrow_column = []

    # Create a table
    rows, cols = data.shape
    table = Table(ax, bbox=[0, 0, 1, 1])

    # Calculate cell height and width
    height = cell_height
    width = 1.0 / cols

    # Add header cells
    for j, col in enumerate(data.columns):
        table.add_cell(0, j, width, height,
                       text=col,
                       loc='center',
                       facecolor='lightgrey')

    # Normalize data for coloring (by column, excluding non-numeric)
    numeric_cols = data.select_dtypes(include=np.number).columns
    norm_data = pd.DataFrame(index=data.index, columns=data.columns)

    for col in numeric_cols:
        if col in data.columns:
            col_min, col_max = data[col].min(), data[col].max()
            if col_max > col_min:  # Avoid division by zero
                norm_data[col] = (data[col] - col_min) / (col_max - col_min)
            else:
                norm_data[col] = 0.5  # Set a neutral value if all values are the same

    # Add data cells with heatmap coloring
    cmap_func = plt.cm.get_cmap(cmap)

    for i, (idx, row) in enumerate(data.iterrows(), 1):
        # Add row header
        table.add_cell(i, -1, width, height,
                       text=str(idx),
                       loc='center',
                       facecolor='lightgrey')

        for j, col in enumerate(data.columns):
            cell_value = row[col]
            cell_text = str(cell_value)

            # Determine cell color based on normalized value (if numeric)
            if col in numeric_cols:
                norm_value = norm_data.loc[idx, col]
                if not np.isnan(norm_value):
                    bg_color = cmap_func(norm_value)
                else:
                    bg_color = 'white'
            else:
                bg_color = 'white'

            # Add the cell
            table.add_cell(i, j, width, height,
                          text=cell_text,
                          loc='center',
                          facecolor=bg_color)

            # Add arrow if this column should have direction indicators
            if col in arrow_column and col in numeric_cols:
                try:
                    if i > 1:  # Skip the first row as we need a previous value to compare
                        prev_value = data.iloc[i-2][col]  # i-2 because i starts at 1 and we need to offset for header
                        curr_value = cell_value

                        # Determine arrow direction
                        if curr_value > prev_value:
                            arrow_dir = "↑"  # Up arrow
                            arrow_color = 'green'
                        elif curr_value < prev_value:
                            arrow_dir = "↓"  # Down arrow
                            arrow_color = 'red'
                        else:
                            arrow_dir = "→"  # Right arrow (no change)
                            arrow_color = 'gray'

                        # Add arrow annotation
                        cell = table[i, j]
                        text = cell.get_text()
                        text.set_text(f"{cell_text} {arrow_dir}")
                        text.set_color('black')
                        text.set_fontweight('bold')
                except:
                    pass  # Skip if there's an error with the arrow

    # Add the table to the plot
    ax.add_table(table)

    # Add title if provided
    if title:
        ax.set_title(title, pad=20, fontsize=font_size+4)

    # Adjust layout
    plt.tight_layout()

    return fig, ax

# Example usage
if __name__ == "__main__":
    # Create sample data
    np.random.seed(42)
    dates = pd.date_range(start='2023-01-01', periods=10, freq='M')

    data = {
        'Revenue': np.random.randint(100, 200, 10),
        'Expenses': np.random.randint(70, 150, 10),
        'Profit': np.random.randint(10, 70, 10),
        'Growth (%)': np.round(np.random.uniform(-5, 15, 10), 1),
        'Region': np.random.choice(['North', 'South', 'East', 'West'], 10)
    }

    # Calculate some derived values for more realistic data
    data['Profit'] = [r - e for r, e in zip(data['Revenue'], data['Expenses'])]

    # Create DataFrame
    df = pd.DataFrame(data, index=dates)

    # Create and display the table with arrows on the 'Growth (%)' column
    fig, ax = create_table_with_heatmap_arrows(
        df,
        title='Monthly Financial Performance',
        arrow_column=['Growth (%)', 'Profit'],
        figsize=(12, 8)
    )

    plt.savefig('heatmap_table_with_arrows.png', dpi=300, bbox_inches='tight')
    plt.show()

# Example 2: More advanced styling
def advanced_styled_table_with_arrows(data, title=None, cmap='RdYlGn_r',
                                      trend_columns=None, figsize=(12, 8)):
    """
    Create a more professionally styled table with trend arrows and heatmap coloring.

    Parameters:
    -----------
    data : pandas DataFrame
        The data to visualize in the table.
    title : str, optional
        Title of the table.
    cmap : str or matplotlib colormap, optional
        Colormap to use for the heatmap.
    trend_columns : list, optional
        Columns to add trend indicators.
    figsize : tuple, optional
        Figure size (width, height).
    """
    # Create a figure and axis
    fig, ax = plt.subplots(figsize=figsize)
    ax.set_axis_off()

    # Set up the table
    nrows, ncols = data.shape

    # Calculate positions for cells
    cell_width = 1.0 / (ncols + 1)  # +1 for row headers
    cell_height = 0.8 / (nrows + 1)  # +1 for column headers

    # Convert cmap to colormap function
    cmap_func = plt.cm.get_cmap(cmap)

    # Normalize data for coloring
    numeric_cols = data.select_dtypes(include=np.number).columns
    norm_by_col = {}

    for col in numeric_cols:
        col_min, col_max = data[col].min(), data[col].max()
        if col_max > col_min:  # Avoid division by zero
            norm_by_col[col] = mcolors.Normalize(vmin=col_min, vmax=col_max)
        else:
            norm_by_col[col] = mcolors.Normalize(vmin=col_min-1, vmax=col_max+1)

    # Draw column headers
    header_bg = '#f0f0f0'
    for j, col in enumerate(data.columns):
        x = (j + 1) * cell_width
        rect = plt.Rectangle((x, 1-cell_height), cell_width, cell_height,
                             fill=True, color=header_bg, ec='white', lw=2, zorder=2)
        ax.add_patch(rect)
        ax.text(x + cell_width/2, 1-cell_height/2, col, ha='center', va='center',
                fontweight='bold', fontsize=11, zorder=3)

    # Draw row headers
    for i, idx in enumerate(data.index):
        y = 1 - ((i + 2) * cell_height)
        rect = plt.Rectangle((0, y), cell_width, cell_height,
                            fill=True, color=header_bg, ec='white', lw=2, zorder=2)
        ax.add_patch(rect)
        ax.text(cell_width/2, y + cell_height/2, str(idx), ha='center', va='center',
                fontweight='bold', fontsize=10, zorder=3)

    # Draw data cells
    for i, idx in enumerate(data.index):
        row_y = 1 - ((i + 2) * cell_height)

        for j, col in enumerate(data.columns):
            col_x = (j + 1) * cell_width
            value = data.loc[idx, col]

            # Determine cell color
            if col in numeric_cols:
                norm = norm_by_col[col]
                cell_color = cmap_func(norm(value))
            else:
                cell_color = 'white'

            # Create cell
            rect = plt.Rectangle((col_x, row_y), cell_width, cell_height,
                                fill=True, color=cell_color, ec='white', lw=2, zorder=2)
            ax.add_patch(rect)

            # Format value for display
            if isinstance(value, (int, float)):
                if value % 1 == 0:  # Integer
                    display_val = f"{int(value):,}"
                else:  # Float
                    display_val = f"{value:.1f}"

                # Add % sign if the column name suggests percentages
                if 'percent' in col.lower() or '%' in col:
                    display_val = f"{display_val}%"
            else:
                display_val = str(value)

            # Add value text
            ax.text(col_x + cell_width/2, row_y + cell_height/2, display_val,
                    ha='center', va='center', fontsize=10, zorder=3)

            # Add trend arrow if applicable
            if trend_columns and col in trend_columns and i > 0:
                prev_value = data.iloc[i-1][col]
                curr_value = value

                # Determine arrow properties based on trend
                if curr_value > prev_value:
                    arrow_color = 'green'
                    arrow_char = '▲'
                    change_pct = ((curr_value - prev_value) / abs(prev_value)) * 100 if prev_value != 0 else 100
                    tooltip = f"+{change_pct:.1f}%"
                elif curr_value < prev_value:
                    arrow_color = 'red'
                    arrow_char = '▼'
                    change_pct = ((prev_value - curr_value) / abs(prev_value)) * 100 if prev_value != 0 else 100
                    tooltip = f"-{change_pct:.1f}%"
                else:
                    arrow_color = 'gray'
                    arrow_char = '►'
                    tooltip = "No change"

                # Add the arrow annotation
                arrow_x = col_x + cell_width * 0.8
                arrow_y = row_y + cell_height * 0.5
                ax.text(arrow_x, arrow_y, arrow_char, color=arrow_color,
                        fontsize=12, ha='center', va='center', fontweight='bold', zorder=4)

    # Add title if provided
    if title:
        plt.suptitle(title, fontsize=14, fontweight='bold', y=0.98)

    # Set the axis limits
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)

    plt.tight_layout()
    return fig, ax

# Example usage of the advanced styled table
if __name__ == "__main__":
    # Create more complex sample data
    np.random.seed(123)
    dates = pd.date_range(start='2023-01-01', periods=12, freq='M')

    # More realistic financial data
    revenue = [100 + i*5 + np.random.randint(-10, 15) for i in range(12)]
    expenses = [70 + i*2 + np.random.randint(-8, 10) for i in range(12)]
    profit = [r - e for r, e in zip(revenue, expenses)]

    # Calculate growth rates
    revenue_growth = [0]
    for i in range(1, 12):
        growth = ((revenue[i] - revenue[i-1]) / revenue[i-1]) * 100
        revenue_growth.append(round(growth, 1))

    # Create DataFrame
    df_advanced = pd.DataFrame({
        'Revenue ($K)': revenue,
        'Expenses ($K)': expenses,
        'Profit ($K)': profit,
        'Growth (%)': revenue_growth,
        'Performance': ['Good' if p > 25 else 'Average' if p > 15 else 'Poor' for p in profit]
    }, index=dates)

    # Create the advanced table
    fig, ax = advanced_styled_table_with_arrows(
        df_advanced,
        title='Monthly Financial Performance (2023)',
        trend_columns=['Revenue ($K)', 'Profit ($K)', 'Growth (%)'],
        figsize=(12, 8)
    )

    plt.savefig('advanced_heatmap_table.png', dpi=300, bbox_inches='tight')
    plt.show()
```
